```python
def get_realtime_btc_price(symbol):
    # Simbol untuk Bitcoin di Yahoo Finance adalah BTC-USD
    #symbol = 'BTC-USD'

    # Ambil data real-time menggunakan yf.download
    data = yf.download(symbol, period='1d', interval='1m')

    # Dapatkan harga terbaru (price) dari data yang diambil
    latest_price = data['Close'][-1]

    return latest_price
```

```python
def do_forecast_bitcoin():
    # Ambil tanggal hari ini
    today_date = datetime.now()

    # Ambil tanggal esok hari
    esok_date = today_date + timedelta(days=1)
    esok_date_str = esok_date.strftime('%Y-%m-%d')

    # Unduh data harga Bitcoin dari '2022-01-01' hingga 'esok_date_str'
    bitcoin_quote = yf.download('BTC-USD', start='2022-01-01', end=esok_date_str)

    # Filter kolom 'Close' untuk mendapatkan harga penutupan harian
    new_df = bitcoin_quote.filter(['Close'])

    # Lakukan scaling pada data harga
    last_60_days_scaled = scaler.transform(new_df[-60:].values)

    # Siapkan data untuk prediksi
    X_test = np.array([last_60_days_scaled])

    # Lakukan prediksi harga Bitcoin
    pred_price = model.predict(np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1)))

    # Invers scaling pada hasil prediksi
    pred_price = scaler.inverse_transform(pred_price)

    return pred_price[0][0]
```

```python
def chatbot_lstm(update: Update, context: CallbackContext) -> None:
  texts_p = []
  prediction_input = update.message.text
  print(prediction_input)
  # Menghapus punktuasi dan konversi ke huruf kecil
  prediction_input = [letters.lower() for letters in prediction_input if letters not in string.punctuation]
  prediction_input = ''.join(prediction_input)
  texts_p.append(prediction_input)

  # Tokenisasi dan Padding
  prediction_input = tokenizer.texts_to_sequences(texts_p)
  prediction_input = np.array(prediction_input).reshape(-1)
  prediction_input = pad_sequences([prediction_input],15)

  # Mendapatkan hasil keluaran pada model
  output = modelchatbot.predict(prediction_input)
  output = output.argmax()

  # Menemukan respon sesuai data tag dan memainkan voice bot
  response_tag = le.inverse_transform([output])[0]
  if response_tag in responses and responses[response_tag] is not None:
    response = responses[response_tag]
    if(isinstance(response, list)):
      update.message.reply_text(random.choice(response))
    else:
      update.message.reply_text(response)
  else:
    update.message.reply_text("Mohon maaf kami tidak memiliki pengetahuan tentang itu mohon tanya pertanyaan lain:")
    print("="*60 + "\n")
```
